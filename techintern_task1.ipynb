```python
plt.figure(figsize=(10, 6))
products_df['product_category_name'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Product Counts in Each Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='product_length_cm', y='product_width_cm', data=products_df, color='green')
plt.title('Product Length vs Width')
plt.xlabel('Product Length (cm)')
plt.ylabel('Product Width (cm)')
plt.show()
```

```python
# Bar plot - Top 10 cities with the most customers
plt.figure(figsize=(10, 6))
top_cities = customer_df['customer_city'].value_counts().head(10)
top_cities.plot(kind='barh', color='salmon')
plt.title('Top 10 Cities with the Most Customers')
plt.xlabel('Number of Customers')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
# Histogram - Distribution of product prices
plt.figure(figsize=(7, 6))
sns.histplot(orditems_df['price'], bins=40, kde=True, color='skyblue')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
