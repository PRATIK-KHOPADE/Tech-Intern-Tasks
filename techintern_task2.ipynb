```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/Users/91959/Downloads/fifa_eda_stats.csv')
```

```python
#delete duplicate rows
df.drop_duplicates()
```

```python
df.describe
```

```python
# Clean € cells by removing special characters
df['Value'] = df['Value'].str.replace('€', '')
df['Wage'] = df['Wage'].str.replace('€', '')
df['Release Clause'] = df['Release Clause'].str.replace('€', '')

```

```python
df.Value
```

```python
# Convert values to numeric
df['Value'] = df['Value'].str.replace('K', 'e3').str.replace('M', 'e6').astype(float)
df['Wage'] = df['Wage'].str.replace('K', 'e3').str.replace('M', 'e6').astype(float)
df['Release Clause'] = df['Release Clause'].str.replace('K', 'e3').str.replace('M', 'e6').astype(float)
```

```python
# Drop players with no value
df = df[df['Value'] != 0]
```

```python
# Visualizations - Example: Histogram of player ages
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example: Scatter plot between Overall Rating and Potential
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Overall', y='Potential', data=df, hue='Preferred Foot')
plt.title('Overall Rating vs Potential by Preferred Foot')
plt.xlabel('Overall Rating')
plt.ylabel('Potential')
plt.legend(title='Preferred Foot')
plt.show()
```

```python
# Example: Bar plot of top 10 nationalities by count
top_nationalities = df['Nationality'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_nationalities.index, y=top_nationalities.values)
plt.title('Top 10 Nationalities of Players')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Simulated data for team rankings across seasons
seasons = ['Season 1', 'Season 2', 'Season 3', 'Season 4']
teams = ['Team A', 'Team B', 'Team C', 'Team D']
ranking_data = []

for season in seasons:
    for team in teams:
        ranking_data.append({
            'Season': season,
            'Team': team,
            'Ranking': random.randint(1, len(teams))  # Simulated ranking
        })

# Convert the data to a DataFrame
ranking_df = pd.DataFrame(ranking_data)

# Bar chart to compare team rankings across seasons
plt.figure(figsize=(10, 6))
colors = plt.cm.get_cmap('tab10', len(teams))

for i, team in enumerate(teams):
    team_data = ranking_df[ranking_df['Team'] == team]
    plt.bar(team_data['Season'], team_data['Ranking'], color=colors(i), label=team)

plt.title('Team Rankings Across Seasons')
plt.xlabel('Season')
plt.ylabel('Ranking')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Simulated data for player performance over seasons
# Create a DataFrame with columns: Season, Player, Performance
seasons = ['Season 1', 'Season 2', 'Season 3', 'Season 4']
players = ['Player A', 'Player B', 'Player C', 'Player D']
performance_data = []

for season in seasons:
    for player in players:
        performance_data.append({
            'Season': season,
            'Player': player,
            'Performance': random.randint(60, 100)  # Simulated performance score
        })

# Convert the data to a DataFrame
performance_df = pd.DataFrame(performance_data)

# Line plot to visualize player performance over seasons
plt.figure(figsize=(10, 6))
for player in players:
    player_data = performance_df[performance_df['Player'] == player]
    plt.plot(player_data['Season'], player_data['Performance'], marker='o', label=player)

plt.title('Player Performance Over Seasons')
plt.xlabel('Season')
plt.ylabel('Performance')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
