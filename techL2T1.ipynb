```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('C:/Users/91959/Documents/MovieDataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
display(df.describe())
```

```python
# Visualize the distribution of vote_average
plt.figure(figsize=(10, 6))
sns.histplot(df['vote_average'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Vote Average')
plt.xlabel('Vote Average')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the relationship between popularity and vote_average
plt.figure(figsize=(10, 6))
sns.scatterplot(x='popularity', y='vote_average', data=df, alpha=0.5)
plt.title('Scatter Plot: Popularity vs. Vote Average')
plt.xlabel('Popularity')
plt.ylabel('Vote Average')
plt.show()
```

```python
# Explore movie release patterns over time
df['release_date'] = pd.to_datetime(df['release_date'])
df['release_year'] = df['release_date'].dt.year

plt.figure(figsize=(12, 6))
sns.countplot(x='release_year', data=df, palette='viridis')
plt.title('Number of Movies Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the distribution of 'popularity' using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['popularity'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the relationship between 'vote_average' and 'popularity' using a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x='vote_average', y='popularity', data=df, alpha=0.5)
plt.title('Scatter Plot: Vote Average vs. Popularity')
plt.xlabel('Vote Average')
plt.ylabel('Popularity')
plt.show()
```

```python
# Explore the relationship between 'vote_count' and 'popularity' using a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x='vote_count', y='popularity', data=df, alpha=0.5)
plt.title('Scatter Plot: Vote Count vs. Popularity')
plt.xlabel('Vote Count')
plt.ylabel('Popularity')
plt.show()
```

```python
dff= pd.read_csv('C:/Users/91959/Downloads/IMDB.csv',encoding='ISO-8859-1')
```

```python
dff.info()
```

```python
# Create a pivot table to aggregate average ratings for each genre
genre_rating_pivot = dff.pivot_table(index='Genre', values='Rating', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(20, 8))
sns.heatmap(genre_rating_pivot, cmap='YlGnBu', annot=True, fmt='.2f', linewidths=.5)
plt.title('Average Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.show()
```

```python
# Plot histogram of Ratings
plt.figure(figsize=(10, 6))
sns.histplot(dff['Rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
